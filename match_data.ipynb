```python
for i in range(len(detail_df)):
    detail_search_df = df[(df["父牛"]==detail_df["父牛"][i]) & (df["母の父"]==detail_df["母の父"][i]) & (df["母の祖父"]==detail_df["母の祖父"][i]) & (df["母の祖祖父"]==detail_df["母の祖祖父"][i]) & (df["日令"]==detail_df["日令"][i]) & (df["性別"]==detail_df["性別"][i]) & (df["年月日"]==detail_df["年月日"][i])]
    if len(detail_search_df) == 1:
        detail_numpy = detail_df.to_numpy()
        detail_part = detail_numpy[[i]]
        serch_numpy = detail_search_df.to_numpy()
        if i == 0:
            farst_numpy = np.hstack((detail_part,serch_numpy))
        else:
            second_numpy = np.hstack((detail_part,serch_numpy))
            farst_numpy = np.vstack((farst_numpy, second_numpy))
farst_numpy
```
